```python
import numpy as np
import pandas as pd
import random as rd
import os


data = pd.read_csv('dataset_sonidos_peligrosos_recortado.csv')

print(data.info())
valores = data['src_file']

100032 in valores.values
```

```python
def generar_numers_random():
    valor = rd.randint(1000, 999999)
    return valor if valor not in valores.values else generar_numers_random()

def generar_numers_fold(valor, cantidad_archivos):
    valordividir = cantidad_archivos//5
    if valor>0 and valor<=valordividir:
        return 1
    elif valor>8 and valor<=valordividir*2:
        return 2 
    elif valor>16 and valor<=valordividir*3:
        return 3
    elif valor>24 and valor<=valordividir*4:
        return 4
    else:
        return 5

print(generar_numers_random())
```

```python
#funcion verificar existencia de ruta
def verificar_ruta(ruta):
    if os.path.exists(ruta):
           # Accedemos al directorio
        os.chdir(ruta)
        print(f'Ahora estás en el directorio: {os.getcwd()}')
        return True
    else:
        print(f'El directorio no existe: {ruta}')
        return False
```

- df.insert(): Modifica el DataFrame original directamente, no crea una copia.
- df.append(): Crea un nuevo DataFrame una copia con la fila adicional y debe asignarse a una variable si quieres conservarlo.

```python

# Supongamos que tienes un DataFrame existente
data = {'filename': [],
        'fold': [],
        'target': [],
        'category': [],
        'esc10': [],
        'src_file': [],
        'take': []}

df = pd.DataFrame(data)

# Creamos una nueva fila como un diccionario
filas = []
targeta = 9
for i in range(1, cantidad_archivos+1):
    # Generar un nuevo número aleatorio para la siguiente iteración
    numero_aleatorio = generar_numers_random()
    numero_fold = generar_numers_fold(i, cantidad_archivos)
    fila = {'filename': str(numero_fold)+'-'+str(numero_aleatorio)+'-'+'A-'+str(targeta)+'.wav', 
            'fold': numero_fold, 
            'target': targeta,
            'category': 'tormenta_electrica', 
            'esc10': 'False', 
            'src_file': numero_aleatorio, 
            'take': 'A'}
    # Nombre antiguo y nuevo del archivo
    antiguo_nombre = f'tormenta_electrica({i}).wav'
    nuevo_nombre = fila['filename']
    # Renombrar el archivo
    os.rename(os.path.join(ruta_directorio_corriente_electrica_fallando, antiguo_nombre), os.path.join(ruta_directorio_corriente_electrica_fallando_v1, nuevo_nombre))
    filas.append(fila)


# Usamos append para agregar la nueva fila al DataFrame
for i in filas:
    df = df.append(i, ignore_index=True)



```
