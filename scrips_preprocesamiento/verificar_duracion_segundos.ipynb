```python
import os
import librosa

# Ruta de la carpeta con los archivos de audio
carpeta = 'D:/SEMESTRE 2-2023/INTELIGENCIA ARTIFICIAL 3 - SIS330/modelos_deteccion_sonidos/dataset_sonidos_peligrosos'

# Lista para almacenar los nombres de los archivos que cumplen con los criterios
archivos_menos_5s = []
archivos_mas_5s = []

# Iterar a través de los archivos en la carpeta
for archivo in os.listdir(carpeta):
    if archivo.endswith('.wav'):  # Asumiendo que los archivos son en formato WAV
        ruta_archivo = os.path.join(carpeta, archivo)
        duracion = librosa.get_duration(filename=ruta_archivo)
        
        if duracion == 30:
            archivos_menos_5s.append(archivo)
        else:
            archivos_mas_5s.append(archivo)

# Imprimir los resultados
print("Archivos con menos de 30 segundos:")
for archivo in archivos_menos_5s:
    print(archivo)

print("\nArchivos con 30 segundos o más:")
for archivo in archivos_mas_5s:
    print(archivo)

print(len(archivos_menos_5s))
```

```python
from pydub import AudioSegment
import os
import librosa

carpeta_cortados_may_5s = 'D:/SEMESTRE 2-2023/INTELIGENCIA ARTIFICIAL 3 - SIS330/modelos_deteccion_sonidos/dataset_sonidos_peligrosos_10seg'
if not os.path.exists(carpeta_cortados_may_5s):
    os.makedirs(carpeta_cortados_may_5s)

# Iterar a través de los archivos en la carpeta
for archivo in os.listdir(carpeta):
    if archivo.endswith('.wav'):  # Asumiendo que los archivos son en formato WAV
        ruta_archivo = os.path.join(carpeta, archivo)
        duracion = librosa.get_duration(filename=ruta_archivo)
        
        if duracion > 10:# 30 segundos = 30000 milisegundos
            # Cargar el archivo de audio
            audio = AudioSegment.from_file(ruta_archivo, format="wav")
            
            # Cortar el audio a 30 segundos
            audio_cortado = audio[:10000]  # 30 segundos = 30000 milisegundos
            
            # Guardar el archivo cortado en la carpeta de archivos cortados
            ruta_archivo_cortado = os.path.join(carpeta_cortados_may_5s, archivo)
            audio_cortado.export(ruta_archivo_cortado, format="wav")
            
            # Reemplazar el archivo original con el archivo cortado
            os.remove(ruta_archivo)
            os.rename(ruta_archivo_cortado, ruta_archivo)
```
