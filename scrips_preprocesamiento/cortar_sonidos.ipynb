```python
from pydub import AudioSegment
import os

def detect_loud_part(audio_segment, threshold=-20):
    """
    Detecta el punto donde el audio supera un cierto umbral de volumen.
    Retorna la posición en milisegundos.
    """
    for ms in range(len(audio_segment)):
        if audio_segment[ms].dBFS > threshold:
            return ms
    return 0

# Directorio donde están tus archivos .wav
directory = "D:/SEMESTRE 2-2023/INTELIGENCIA ARTIFICIAL 3 - SIS330/modelos_deteccion_sonidos/HTS-Audio-Transformer/workspace/dataset_sonidos_peligrosos/raw/sonidos_peligrosos/audio"

# Procesa cada archivo en el directorio
for filename in os.listdir(directory):
    if filename.endswith(".wav"):
        path = os.path.join(directory, filename)
        audio = AudioSegment.from_wav(path)

        # Encuentra el punto de inicio del ruido
        start_ms = detect_loud_part(audio)

        # Recorta 10 segundos a partir del punto de inicio
        end_ms = start_ms + 5000  # 5000 ms = 5 s
        trimmed_audio = audio[start_ms:end_ms]

        # Guarda el archivo recortado
        trimmed_audio.export(path, format="wav")

print("Proceso completado.")
```

```python
import librosa
import soundfile as sf
import os

def resample_audio(input_folder, output_folder, target_sr):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for filename in os.listdir(input_folder):
        if filename.endswith('.wav'):
            file_path = os.path.join(input_folder, filename)
            y, _ = librosa.load(file_path, sr=target_sr)  # Re-muestrear
            output_file_path = os.path.join(output_folder, filename)
            sf.write(output_file_path, y, target_sr)  # Guardar archivo re-muestreado

# Configura las carpetas y la tasa de muestreo deseada
input_folder = 'G:/Mi unidad/DatasetAudiosPrueba/audio/audio/sonidos_peligrosos_recortado'
output_folder = 'G:/Mi unidad/dataset_sonidos_peligrosos_recortado_completo/audio'
target_sr = 44100 

resample_audio(input_folder, output_folder, target_sr)
```
