```python
import librosa
from scipy.io.wavfile import write
import numpy as np

#directorioAudiosCortados = 'D:/SEMESTRE 2-2023/INTELIGENCIA ARTIFICIAL 3 - SIS330/datasetSonidos/ruta_salida_2/'
# input_path = ruta_directorio+'1-497959-A-51.wav'
# output_path = ruta_salida_2+'1-497959-A-51.wav'

for file in os.listdir(directorio):
    input_path = directorio+file
    output_path = ruta_salida_2+file
    # Cargar el archivo de audio
    y, sr = librosa.load(input_path)

    # Convertir a 16 bits
    y_16bits = (y * 32767).astype(np.int16)

    # Guardar el nuevo archivo de audio
    write(output_path, sr, y_16bits)
```
